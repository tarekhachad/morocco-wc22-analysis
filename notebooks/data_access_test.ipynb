 ### Confirm WC2022 is available

```python
from statsbombpy import sb

compets = sb.competitions()
wc2022 = compets[
    (compets['competition_name'] == 'FIFA World Cup') &
    (compets['season_name'] == '2022')
]
print(wc2022[['competition_id', 'season_id', 'competition_name', 'season_name']])
```

### Confirm Morocco matches are available

```python
matches = sb.matches(competition_id=43, season_id=106)

morocco = matches[
    (matches['home_team'] == 'Morocco') |
    (matches['away_team'] == 'Morocco')
]

print(morocco[['match_id', 'home_team', 'away_team', 'home_score', 'away_score']])
```
